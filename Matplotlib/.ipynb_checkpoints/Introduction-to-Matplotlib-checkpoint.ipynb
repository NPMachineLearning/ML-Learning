### Matplotlib workflow

```python
# setp 1 import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# step 2 prepare data
x = [1,2,3]
y = [11, 20, 31]

# step 3 setup figure
fig, ax = plt.subplots(figsize=(10,10))

# step 4 plot
ax.plot(x,y)

#step 5 customize plot
ax.set(title='example plot', xlabel='x-axis', ylabel='y-axis')

#step 6 save & show
fig.savefig('sample-plot.png')
```

### Make plot with Numpy array
* line plot
* scatter plot

```python
# import numpy
import numpy as np
```

```python
# create some data
x = np.linspace(0, 10, 100)
```

```python
x[:20]
```

```python
# line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# bar plot
butter_prices = {
    "Peanut Butter": 10,
    "Almond Butter": 15,
    "Cashew Butter": 12
}
fig, ax = plt.subplots()
ax.bar(butter_prices.keys(), butter_prices.values());
ax.set(title='Butter Store', xlabel='(Butter types)', ylabel='(Prices $)');
```

```python
# bar horizontal plot
fig, ax = plt.subplots()
ax.barh(list(butter_prices.keys()), list(butter_prices.values()));
```

```python
# histogram plot
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options subplots

```python
# subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax1.plot(x, x/2);
ax2.scatter(np.random.rand(100), np.random.rand(100));
ax3.bar(butter_prices.keys(), butter_prices.values());
ax4.hist(x);
```

```python
# subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.rand(100), np.random.rand(100));
ax[1,0].bar(butter_prices.keys(), butter_prices.values());
ax[1,1].hist(x);
```

### Which one to use pyplot or matplotlib OO method?
* pyplot: When you need to quick glance the plot result from data
* OO method: When you need advance plot

```python
# pyplot
heart_disease_over_50 = heart_disease[heart_disease['age'] > 50]
heart_disease_over_50.head()
```

```python
heart_disease_over_50.plot(kind='scatter', x='age', y='chol', c='target', figsize=(10,6));
```

```python
# OO method mix with pyplot
fig, ax = plt.subplots(figsize=(10,6))

#set x axis value range from 45 ~ 100
#ax.set_xlim([45, 100])

heart_disease_over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);
```

```python
## OO method
fig, ax = plt.subplots(figsize=(10,6))
# draw scatter graphics
scatter = ax.scatter(x=heart_disease_over_50['age'],
           y=heart_disease_over_50['chol'],
           c=heart_disease_over_50['target']
          );
# customize
ax.set(xlabel='Age', ylabel='Cholesterol', title='Heart Disease with Cholesterol Level');
ax.legend(*scatter.legend_elements(), title='Target');
ax.axhline(heart_disease_over_50['chol'].mean(), linestyle='-.');
```

```python
# legend
legend_scatter = plt.scatter([1,2,3], [4,5,6], c=[7,8,9])
plt.legend(*legend_scatter.legend_elements());
```

```python
heart_disease_over_50.head()
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(
    nrows=3, 
    ncols=1, 
    figsize=(10, 10),
    sharex=True
)

# add data to ax0
ax0_scatter = ax0.scatter(
    x=heart_disease_over_50['age'], 
    y=heart_disease_over_50['chol'], 
    c=heart_disease_over_50['target']
);

# customzie ax0
ax0.set(
    title='Heart Disease Cholesterol', 
#     xlabel='Age', 
    ylabel='Cholesterol'
);
ax0.legend(*ax0_scatter.legend_elements(), title='Target');
ax0.axhline(y=heart_disease_over_50['chol'].mean(), linestyle='--');

# addd data to ax1
ax1_scatter = ax1.scatter(
    x=heart_disease_over_50['age'], 
    y=heart_disease_over_50['thalach'], 
    c=heart_disease_over_50['target']
);

#customzie ax1
ax1.set(
    title='Heart Disease Thalach', 
#     xlabel='Age', 
    ylabel='Thalach (Max heart rate)'
);
ax1.legend(*ax1_scatter.legend_elements(), title='Target');
ax1.axhline(y=heart_disease_over_50['thalach'].mean(), linestyle='--');

# add data to ax2
ax2_bar = ax2.bar(
    x=heart_disease_over_50['age'], 
    height=heart_disease_over_50['trestbps']
);

# customzie ax2
ax2.set(title='Heart Disease Trestbps', xlabel='Age', ylabel='Trestbps');
ax2.axhline(y=heart_disease_over_50['trestbps'].mean(), linestyle='--');

#fig title
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```

```python
plt.style.use('seaborn-whitegrid')

fig, (ax0, ax1) = plt.subplots(
    nrows=2, 
    ncols=1, 
    figsize=(10, 10),
    sharex=True
)

# add data to ax0
ax0_scatter = ax0.scatter(
    x=heart_disease_over_50['age'], 
    y=heart_disease_over_50['chol'], 
    c=heart_disease_over_50['target'],
    cmap='winter'
);

# customzie ax0
ax0.set(
    title='Heart Disease Cholesterol', 
#     xlabel='Age', 
    ylabel='Cholesterol'
);
ax0.legend(*ax0_scatter.legend_elements(), title='Target');
ax0.axhline(y=heart_disease_over_50['chol'].mean(), linestyle='--');
ax0.set_xlim([50,80]);

# addd data to ax1
ax1_scatter = ax1.scatter(
    x=heart_disease_over_50['age'], 
    y=heart_disease_over_50['thalach'], 
    c=heart_disease_over_50['target'],
    cmap='winter'
);

#customzie ax1
ax1.set(
    title='Heart Disease Thalach', 
#     xlabel='Age', 
    ylabel='Thalach (Max heart rate)'
);
ax1.legend(*ax1_scatter.legend_elements(), title='Target');
ax1.axhline(y=heart_disease_over_50['thalach'].mean(), linestyle='--');
ax1.set_ylim([60,200]);
```
